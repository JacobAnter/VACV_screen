```python
"""
The purpose of this Jupyter notebook is to generate a FASTA file for the
validation set as well as for the test set.
"""
```

```python
import os

import pandas as pd
from biotite.sequence.io import fasta
```

### Loading Files

```python
# Load the validation set TSV file storing the PPI pairs
data_set_dir = "data_set_files"

val_set_path = os.path.join(
    data_set_dir,
    "bullet-proof_validation_set_without_header.tsv"
)

val_set_df = pd.read_csv(
    val_set_path,
    sep="\t",
    header=None
)
```

```python
# Load the test set TSV file storing the PPI pairs
test_set_path = os.path.join(
    data_set_dir,
    "bullet-proof_test_set_without_header.tsv"
)

test_set_df = pd.read_csv(
    test_set_path,
    sep="\t",
    header=None
)
```

```python
# Also load the FASTA file storing all protein sequences
all_prots_fasta_path = os.path.join(
    data_set_dir,
    "bullet-proof_data_set.fasta"
)

all_prots_fasta = fasta.FastaFile.read(all_prots_fasta_path)
```

### Generating a FASTA File for the Validation Set

```python
# Determine the unique UniProt accessions comprised in the validation
# set
unique_val_prots = val_set_df[[0, 1]].stack().unique().tolist()
```

```python
# Now, create a new instance of a FASTA file and populate it with
# sequences of the validation set proteins
val_set_fasta = fasta.FastaFile()

for val_prot in unique_val_prots:
    val_set_fasta[val_prot] = all_prots_fasta[val_prot]
```

```python
assert (
    len(val_set_fasta)
    ==
    len(unique_val_prots)
), (
    "Something went wrong while populating the FASTA file!"
)
```

```python
# Save the FASTA file to disk
val_set_fasta.write(os.path.join(
    data_set_dir,
    "bullet-proof_validation_set.fasta"
))
```

### Generating a FASTA File for the Test Set

```python
# Determine the unique UniProt accessions comprised in the test set
unique_test_prots = test_set_df[[0, 1]].stack().unique().tolist()
```

```python
# Create a new instance of a FASTA file and populate it with sequences
# of the test set proteins
test_set_fasta = fasta.FastaFile()

for test_prot in unique_test_prots:
    test_set_fasta[test_prot] = all_prots_fasta[test_prot]
```

```python
assert (
    len(test_set_fasta)
    ==
    len(unique_test_prots)
), (
    "Something went wrong while populating the FASTA file!"
)
```

```python
# Save the FASTA file to disk
test_set_fasta.write(os.path.join(
    data_set_dir,
    "bullet-proof_test_set.fasta"
))
```
