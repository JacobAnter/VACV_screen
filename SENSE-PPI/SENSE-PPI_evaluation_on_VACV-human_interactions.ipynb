```python
# Owing to memory constraints, SENSE-PPI has not been applied to the
# whole TSV file with the PPI pairs at once but to 17 chunks
# For each of these 17 chunks, a separate results file has been
# generated (strictly speaking, two results files have been generated
# for each chunk, the first of which contains predicted probabilities
# for all PPI pairs in the respective chunk and the second of which
# contains predicted probabilities exclusively for positive PPI pairs)
# Thus, the results files of the individual chunks are stitched
# togehther so as to yield one TSV encompassing all investigated PPI
# pairs and only positive PPI pairs, respectively
```

```python
import pandas as pd
```

```python
# In total, the PPI pairs have been split into 17 chunks with counting
# starting from zero
all_PPIs_probs_df = pd.concat(
    [
        pd.read_csv(
            f"Results_chunk_{i}/predictions_Qiagen_subset_"\
            f"chunk_{i}_size_500,000.tsv",
            sep="\t"
        )
        for i in range(17)
    ],
    ignore_index=True
)
all_PPIs_probs_df.to_csv(
    "predictions_Qiagen_subset.tsv",
    index=False
)

pos_PPIs_probs_df = pd.concat(
    [
        pd.read_csv(
            f"Results_chunk_{i}/predictions_positive_interactions_"\
            f"Qiagen_subset_chunk_{i}_size_500,000.tsv",
            sep="\t"
        )
        for i in range(17)
    ],
    ignore_index=True
)
pos_PPIs_probs_df.to_csv(
    "predictions_positive_interactions_Qiagen_subset.tsv",
    index=False
)
```
